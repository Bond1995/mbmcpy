```python
from scipy import signal
from scipy import linalg
from matplotlib import gridspec

import matplotlib.pyplot as plt
import scipy.io.wavfile
import numpy as np
import os
import math


path_db = '05_AudioClassifier_Pdf/05_AudioClassifier_Pdf/database/'
path_music = path_db+'music/'
path_speech = path_db+'speech/'
path_unknowns = path_db.replace('database/','')+'unknownSounds/'

fig = plt.figure()

for filename in os.listdir(path_music):
    rate,data = scipy.io.wavfile.read(path_music+filename)
    f, t, Sxx = signal.spectrogram(data, rate, 'hamming',512,160,512)
    
    timeArray = np.arange(0, data.shape[0], 1)#arange(start,stop,step)
    timeArray = timeArray / rate
    timeArray = timeArray * 1000  #scale to milliseconds
    gs = gridspec.GridSpec(1,2)
    
    plt.subplot(gs[0,0])
    plt.plot(timeArray, data)
    plt.ylabel('Amplitude')
    plt.xlabel('Time [ms]')
    plt.subplot(gs[0,1])
    plt.pcolormesh(t,f,Sxx)
    plt.xlabel('Frequency [Hz]')
    plt.ylabel('Time [ms]')
    plt.tight_layout()
    plt.show()
```

```python
for filename in os.listdir(path_speech):
    rate,data = scipy.io.wavfile.read(path_speech+filename)
    f, t, Sxx = signal.spectrogram(data, rate, 'hamming',512,160,512)
    
    timeArray = np.arange(0, data.shape[0], 1)#arange(start,stop,step)
    timeArray = timeArray / rate
    timeArray = timeArray * 1000  #scale to milliseconds
    plt.plot(timeArray, data, color='k')
    plt.ylabel('Amplitude')
    plt.xlabel('Time (ms)')
    plt.show()
    
    plt.plot(f,np.sum(Sxx,1))
    plt.xlabel('Frequency [Hz]')
    plt.ylabel('Spectrogram')
    plt.show()
```
